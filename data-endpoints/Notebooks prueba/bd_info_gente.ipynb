```python
import numpy as np

# Insertamos una nueva columna 'ID finca' en la posición 1 en el df
# Los valores se asignan en base al agrupamiento por la columna 'Dirección'
df.insert(1, 'ID finca', df.groupby('Dirección').ngroup() + 1)

for direccion in df['Dirección'].unique():
    codigo_postal = str(np.random.randint(46015, 46026))
    df.loc[df['Dirección'] == direccion, 'Código Postal'] = codigo_postal

# Movemos la columna 'Código Postal' a la posición 4 en el DataFrame
codigo_postal_column = df.pop('Código Postal')
df.insert(4, 'Código Postal', codigo_postal_column)

# Generamos 'Teléfono presidente' para cada 'Presidente finca'
for presidente in df['Presidente finca'].unique():
    telefono_presidente = f'{random.randint(600, 699)}{random.randint(100000, 999999)}'
    df.loc[df['Presidente finca'] == presidente, 'Teléfono presidente'] = telefono_presidente
df
```
