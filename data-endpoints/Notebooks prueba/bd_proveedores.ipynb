```python
import pandas as pd
import numpy as np
from faker import Faker
import random

# Creamos una instancia de Faker 
faker = Faker('es_ES')

# Relación entre servicios y tipos de proveedores
tipos_proveedores = {
    "Mantenimiento áreas comunes": "Mantenimiento",
    "Reparación instalaciones eléctricas": "Electricistas",
    "Fontanería": "Fontaneros",
    "Servicios de limpieza": "Limpieza",
    "Seguridad": "Seguridad",
    "Reparación ascensores": "Mantenimiento",
    "Control de plagas": "Plagas",
    "Pintura y reparación de paredes": "Pintores",
    "Servicios de emergencia": "Emergencia",
    "Mantenimiento calefacción y aire acondicionado": "Mantenimiento",
    "Limpieza y mantenimiento de piscinas": "Piscinas",
    "Gestión de residuos": "Gestión de Residuos"
}

# Lista de servicios
servicios = list(tipos_proveedores.keys())

# Generar 100 filas de datos aleatorios
data = {
    'ID proveedor': list(range(1, 101)),
    'Teléfono': [f"6{random.randint(10000000, 99999999)}" for _ in range(100)],
    'Servicio': [random.choice(servicios) for _ in range(100)],
    'Localidad': ['Valencia' for _ in range(100)],
    'Valoración media': [np.random.uniform(1, 5) for _ in range(100)],
}

data['Nombre proveedor'] = [faker.company_suffix() + " " + tipos_proveedores[s] + " " + faker.last_name() for s in data['Servicio']]

df = pd.DataFrame(data)

nombre_proveedor = df.pop('Nombre proveedor')
df.insert(1, 'Nombre proveedor', nombre_proveedor)

# Imprimir las primeras filas del DataFrame
df
```
